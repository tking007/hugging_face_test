```python
# 使用bash命令执行脚本，安装Miniconda
# -b 选项表示批处理模式，不需要手动确认
# -f 选项表示无错误提示
# -p 选项用于指定安装路径
!bash ./Miniconda3-latest-Linux-x86_64.sh -b -f -p /usr/local
```

```python
# 使用conda命令安装指定版本的Python和ujson库
# -q 选项表示安静模式，减少输出
# -y 选项表示自动确认
# --prefix 选项用于指定安装路径
!conda install -q -y --prefix /usr/local python=3.8.5 ujson
```
